# converting duration in milliseconds to just seconds using lambda expression

```python
d_tracks["duration"]= d_tracks["duration_ms"].apply(lambda x: round(x/1000))
d_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
d_tracks.duration.head()
```

# heatmap using pearson correlation

```python
corr_df = d_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson", numeric_only=True)#drop function will drop unwanted columns(["key", "mode", "explicit"])
plt.figure(figsize=(10,8))
heatmap = sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="mako", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

# To print total no of songs by one particular artist using tracks dataset and displaying the top 10 songs

```python
import pandas as pd
d_tracks = pd.read_csv("C:/tracks.csv")
# Preprocess the "artists" column by converting all text to lowercase and removing leading/trailing spaces
d_tracks['artists'] = d_tracks['artists'].str.lower().str.strip()
# Filtering the DataFrame to get only Justin Bieber songs
bieber_songs = d_tracks[d_tracks['artists'].str.contains('justin bieber', case=False)]
print(bieber_songs['name'])
```

```python
songs = d_tracks[d_tracks['artists'].str.contains('justin bieber', case=False)]
# Sorting the DataFrame based on popularity ( we assume that 'popularity' is a column representing the song's popularity score)
songs_sorted = songs.sort_values(by='popularity', ascending=False)
top_10_recommendations = songs_sorted.head(10)
print(top_10_recommendations[['name', 'popularity']])
```

```python
songs = d_tracks[d_tracks['artists'].str.contains('The Weeknd', case=False)]
print(songs['name'])
#output:length: 144(this prints total no of elements(songs))
```
